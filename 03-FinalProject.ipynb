## Abstract 

This study uses forward selection with a logistic regression model to find what factors most strongly affect employment in the computer science sector. As college students within the tech-sphere ourselves, we sought to discover which factors employers are looking for and conversely, which aren’t as necessary. Internships are becoming increasingly difficult to find, so we wanted to see where, skillwise, the best applicants were strongest to get an idea of what companies generally value. We used a dataset consisting of 70k+ computer science related job applications with select variables including the applicants age, gender, education level, programming languages learned, professional experience, and more. We then applied our logistic regression model on a select set of variables we were most interested in and measured fit using McFadden's Pseudo R-squared. Our findings were that the variable that most likely affected employment were the applicant's computer skills – which indicated how many programming languages the applicant knew. Following this, the amount of professional experience the applicant had, and the education level of the applicant were the following most variables. With this analysis, we then draw conclusions about the future of the computer science job market, and what skills should be honed in on to secure employment.

### Section 1 Exploring Employment Patterns in the Dataset

To identify and compare employment outcomes across key categorical variables, a proportional bar chart is implemented below, comparing education level, gender, and age group. Instead of only showing raw counts, the charts show proportions within each category, which makes comparisons more meaningful when group sizes are uneven.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

cat_cols = ['EdLevel', 'Gender', 'Age']

fig, axes = plt.subplots(1, len(cat_cols), figsize=(20, 8))
fig.suptitle('Employment Rates by Categorical Variables', fontsize=18)

for i, col in enumerate(cat_cols):
    crosstab_prop = pd.crosstab(df[col], df['Employed'], normalize='index')

    if 1 in crosstab_prop.columns:
        crosstab_prop = crosstab_prop.sort_values(by=1, ascending=True)

    crosstab_prop.plot(kind='barh', stacked=True, ax=axes[i], color=['#ff9999', '#66b3ff'], width=0.7)

    axes[i].set_title(f'Proportion of Employed vs.\nUnemployed by {col}', fontsize=14)
    axes[i].set_xlabel('Proportion (0 to 1)')
    axes[i].set_ylabel(col)

    axes[i].legend(title='Status',
                   labels=['0: Unemployed', '1: Employed'],
                   loc='upper center',
                   bbox_to_anchor=(0.5, -0.15),
                   ncol=2)

    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.2f', label_type='center', color='black', weight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

The leftmost chart indicates that, within the dataset, having a higher education level doesn't necessarily correlate with being employed. In fact, this chart reveals that people without any higher education and people with only an undergraduate degree have the highest rates of unemployment.

The middle chart shows us the rates of employment between reported gender groups within our dataset. The chart indicates that men and nonbinary people have a noticeably higher rate of employment than women. It is important to note here, as mentioned above, that our data set is highly skewed towards male respondents (93%), and as such, the gender proportions for women and nonbinary people are much more prone to variance due to smaller sample sizes.

Finally, the rightmost chart shows us that younger applicants (under 35) have a slightly higher employment rate than older applicants. It should be duly noted that the dataset contains more participants under the age of 35 than participants older than 35 (65% and 35%, respectively.)

Overall, these three charts allow us to see existing disparites within the data, relative to each variable.

## ANALYSIS #1 LOGISTIC REGRESSION WITH FORWARD SELECTION
Logistic regression was selected because the target variable in this project, employment status, is binary, indicating whether a respondent was employed or not employed. Since the goal of this analysis is to examine which factors are associated with the likelihood of being employed, logistic regression is appropriate because it models the probability of a binary outcome using multiple predictor variables. In this project, the predictors include technical background, professional experience, education level, age, main professional role, gender, and previous salary. This allows the model to estimate how each factor is related to employment status while accounting for the influence of the other variables in the model.

To identify which variables most strongly predict employment outcome, we first implemented a forward selection model using logistic regression on the full dataset. The variables considered for the model were:-
- Age
- Education Level
- Gender
- MainBranch
- Years of Professional coding experience
- Previous Salary
- Computer Skills

We started with a null model, we iteratively added variables that produced the largest increase in McFadden pseudo R² at each step. Models were ranked by their R² value, and the best-performing model was added permanently before repeating the process.

```python
import statsmodels.formula.api as smf
import numpy as np
import statsmodels as sm    
#Removed Country and YearsCode because they weren't as relevant and Employed since that is the variable we are predicting
columns_of_interest = ['Age', 'C(EdLevel)', 'C(Gender)', 'C(MainBranch)',
       'YearsCodePro', 'PreviousSalary', 'ComputerSkills']
previous_features = [] #start with null model for feature selection
best_r2 = 0.0 # best r-squared seen so far, and by definition a null model has no r squared

while len(columns_of_interest) > 0:
  pseudor2 = [] #a list to keep track of all the models we will try
  for col in columns_of_interest:
      current_features = previous_features + [col]
      indep_vars = ' + '.join(current_features)
      formula = 'Employed ~ {}'.format(indep_vars)
      #disp = 0 also hides print messages from the statsmodel
      a_model = smf.logit(formula, data=df).fit(disp=0) 
      ## To see the full forward selection process uncomment code so it prints at each iteration 
      # print('fitting ', formula, '\t\t R2: {:5.4f}'.format(a_model.prsquared) )
      pseudor2.append( (indep_vars, a_model.prsquared) )

  results = pd.DataFrame.from_records(pseudor2, columns=['independent variables', 'R2']).sort_values(by='R2', ascending=False)

  # here we are setting ourselves up for the next step in model selection
  selected = results['independent variables'].iloc[0] # winning formula is the first one in the sorted R2 list
  previous_features = selected.split(' + ') # remove the formula stuff, leave us with a list of predictors from the winning model of this step
  columns_of_interest.remove( previous_features[-1] ) # remove the new predictor we just added to the formula from the list of predictors to consider next time
  new_r2 = results['R2'].iloc[0] # best pseudo R2 of the step...
  # new_r2 is useful for comparing to best_r2 to see if we've met a stopping condition for forward selection
  #updates new best r2 value
  best_r2 = new_r2 

  ## To see the full forward selection process uncomment code so it prints at each iteration 
  #print('\n on this step we selected the model: {} \t\t R2: {:5.4f}\n vs previous best R2: {:5.4f}'.format(selected, new_r2, best_r2))
  #formula = 'Employed ~ {}'.format(selected)
  #a_model = smf.logit(formula, data=df).fit()
  #print(a_model.summary())

#For notebook clarity, only the final iteration and summary is printed
print('\n on this step we selected the model: {} \t\t R2: {:5.4f}\n vs previous best R2: {:5.4f}'.format(selected, new_r2, best_r2))
formula = 'Employed ~ {}'.format(selected)
a_model = smf.logit(formula, data=df).fit()
print(a_model.summary())
```

To make the full logistic regression results easier to interpret, we converted the model coefficients into odds ratios. Logistic regression coefficients are originally measured in log-odds. By exponentiating each coefficient with np.exp(), we converted them into odds ratios. An odds ratio greater than 1 means the predictor is associated with higher odds of employment, while an odds ratio below 1 means the predictor is associated with lower odds of employment. After making an odds table, a bar chart is used to visualize and compare the odd ratios, The dashed vertical line at 1 represents no change in odds. Bars to the right of 1 indicate higher odds of employment, while bars to the left of 1 indicate lower odds of employment.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Creating Odds Table
params = a_model.params
conf = a_model.conf_int()

odds_table = pd.DataFrame({
    'Variable': params.index,
    'Coefficient': params.values,
    'OddsRatio': np.exp(params.values),
    'LowerCI': np.exp(conf[0].values),
    'UpperCI': np.exp(conf[1].values),
    'PValue': a_model.pvalues.values
})

odds_table = odds_table[odds_table['Variable'] != 'Intercept']

odds_table

# Plot odds

plt.figure(figsize=(10, 7))

sns.barplot(
    data=odds_table.sort_values('OddsRatio'),
    x='OddsRatio',
    y='Variable'
)

plt.axvline(1, linestyle='--')
plt.title('Odds Ratios from Full-Dataset Logistic Regression')
plt.xlabel('Odds Ratio')
plt.ylabel('Predictor')
plt.tight_layout()
plt.show()
```

**Findings**

During forward selection, ComputerSkills produced the largest increase in pseudo R² when added to the model, while the other individual variables produced much smaller improvements. In the final logistic regression model, ComputerSkills had a positive coefficient of approximately 0.327. When converted into an odds ratio, this means that each additional listed computer skill was associated with about 1.39 times higher odds of being employed. When added alone, it increased the model's pseudo R² to around 0.35, while the other variables each produced pseudo R² values close to 0. These findings suggest that the breadth of technical skills is the distinguishing factor between the employed and unemployed applicants within our dataset. 

YearsCodePro was also positively associated with employment, but its effect was much smaller than ComputerSkills. Its coefficient was approximately 0.012, meaning each additional year of professional coding experience was associated with a small increase in the odds of employment. 

MainBranch also had a noticeable relationship with employment. The negative coefficient for NotDev suggests that respondents who were not primarily developers had lower odds of being employed compared to the reference group, This is reasonable because the dataset focuses on Stack Overflow respondents, where technical and developer-related roles are closely tied to employment outcomes.

These results are consistent with the visualizations shown above.

## ANALYSIS #2 GENDER-STRATIFIED LOGISTIC REGRESSION

As previously mentioned the dataset's heavy gender imbalance, with males making up a majority of the sample (93% Male respondents). While we acknowledged this imbalance while doing the full dataset forward selection model, we didn't necessarily address how this may skew results. To supplement the forward selection model on the full dataset, the gender-stratified model is used to analyze whether the full-dataset model reflects patterns in primarily males, or can be applied across different gender groups

To reduce the effects of male-overrepresentation in the model, a balanced dataset is created before running the logistic regression. To do this, we kept all respondents categorized as Woman or NonBinary and then randomly sampled the same number of Man respondents as the combined Woman and NonBinary group. This process is called random undersampling of the majority group.

This approach does not make the dataset perfectly representative of the original population. Instead, it is used as a sensitivity check to reduce the effect of male overrepresentation and allow for a clearer comparison of model results across gender groups.

```python
#Stratified Analysis By Gender
male_df = df[df['Gender'] == 'Man']
female_nb_df = df[df['Gender'].isin(['Woman', 'NonBinary'])]

#Given the overrepresentation of males in the dataset, the stratified analysis on Gender will take a proportionate random sample of males relative to number of female and non-binary respondents
n = len(female_nb_df)
male_sample = male_df.sample(n=n, random_state=42)

balanced_gender_df = pd.concat([male_sample, female_nb_df])
#measures previous salary in tens of thousands for a more informative visualization
balanced_gender_df['PreviousSalary_10k'] = balanced_gender_df['PreviousSalary'] / 10000

balanced_gender_df['Gender'].value_counts()
```

The Man group is still larger than either Woman or NonBinary individually because the male sample was matched to the combined number of Woman and NonBinary respondents, not to the smallest individual group. This choice allowed us to reduce imbalance without unnecessarily removing additional Woman respondents from the analysis.

```python

rows = []

#run logistic regression model on each stratified group to see if there are any noticable difference in valuesss

for gender, group in balanced_gender_df.groupby('Gender'):
    model = smf.logit('Employed ~ ComputerSkills + YearsCodePro + C(EdLevel) + C(Age) + C(MainBranch) + PreviousSalary_10k', data=group).fit(disp=0)
    params = model.params
    conf = model.conf_int()

    for var in ['ComputerSkills', 'YearsCodePro', 'PreviousSalary_10k']:
        rows.append({
            'Gender': gender,
            'Variable': var,
            'OddsRatio': np.exp(params[var]),
            'LowerCI': np.exp(conf.loc[var, 0]),
            'UpperCI': np.exp(conf.loc[var, 1])
        })

    print(f"\n{'='*50}")
    print(f"Gender: {gender}  (n={len(group)})")
    print(f"Pseudo R²: {model.prsquared:.4f}")
    print(model.params.to_string())
```

To make the stratified logistic regression results easier to compare across gender groups, we converted the logistic regression coefficients into odds ratios and plotted the main numerical predictors. Odds ratios above 1 indicate higher odds of employment, while odds ratios below 1 indicate lower odds of employment. This visualization allows us to compare whether predictors such as ComputerSkills, YearsCodePro, and PreviousSalary_10k show similar relationships with employment for Male, Female, and NonBinary respondents.

```python
#Implement plot visualization

#create new table that stores the logistic regression results in a cleaner format
odds_df = pd.DataFrame(rows)

plt.figure(figsize=(9, 6))

sns.pointplot(
    data=odds_df,
    x='Variable',
    y='OddsRatio',
    hue='Gender',
    dodge=True,
    linestyles="None",
)

plt.axhline(1, linestyle='--')
plt.title('Odds Ratios by Gender from Stratified Logistic Regression')
plt.xlabel('Predictor')
plt.ylabel('Odds Ratio')
plt.show()
```

The stratified models showed that ComputerSkills remained a strong positive predictor of employment for all three gender groups. The coefficient for ComputerSkills was approximately 0.307 for Man respondents, 0.338 for NonBinary respondents, and 0.362 for Woman respondents. Converted into odds ratios, this means that each additional listed computer skill was associated with about 1.36 times higher odds of employment for Man respondents, 1.40 times higher odds for NonBinary respondents, and 1.44 times higher odds for Woman respondents.

The pseudo R² values for the stratified models were fairly similar across groups, with values around 0.34 for Man respondents, 0.39 for NonBinary respondents, and 0.38 for Woman respondents. This suggests that the model explained employment outcomes at a somewhat similar level across the three gender groups. Overall, the results of the gender-stratified logistic regression model supports the full-dataset model, where number of computer skills was highly associated with employment even amongst all gender groups. While some coefficients varied slightly between groups, the overall pattern suggests that technical skill breadth was the most stable predictor of employment in both the full model and the stratified models.

## Main Takeaway: Technical Skills Were the Strongest Predictor
The biggest takeaway from our analysis is that in today's computer science job market, the most sought after qualities are practical and technical skills and skill breadth rather than educational credentials and demographic characteristics. ComputerSkills was the strongest predictor both in our visual analysis and logistic regression. Variables such as years of professional coding experience, education level, age and gender did not seem to be a good indicator of employment outcomes. This would suggest that employers place a greater emphasis on current capabilities rather than the amount of time you have spent in the industry. One possible explanation is that due to the rapid advancement in AI and other technologies, the skills required in the industry are constantly changing and those who can adapt to these changes are more likely to remain competitive in the job market. As a result, having a broad and relevant skill set may be more valuable than education level or number of years in the industry. 

## Unexpected Findings and Interpretation
While we didn’t get any results that were inherently contradictory, one result that we  found that didn’t align was the hypothesized “most important” variables and the actual variable with the biggest predictor on employment. While initial visualizations of our data isolated age, gender, and education level, which we proposed had the biggest impact, our forward selection model showed that the amount of computer skills was the strongest predictor of employment, while the other tested variables didn’t seem to have as much impact as a predictor. This factor may simply be a case of not considering self-reporting biases, as applicants who apply for jobs likely inflating the number of computer skills they know, as there isn’t exactly an objective measure of what constitutes as a skill, or how much knowledge is needed to “have” a skill, with these measures being self-reported and therefore prone to bias.

## Limitations of the Dataset and Model
Our work has some limitations, one being that our dataset is a biased sample with 65% being under 35 years old and only 35% being above 35, as well as 93% of the applicants being male and only 5% being female. More data, especially more female applicants would make me a lot more confident, but gender didn’t end up having an effect on our pseudo-R squared when we ran our forward selection with logistic regression, so I don't think it will be that much of an issue. Longitudinal data would be very helpful, as we could track education level, skills gained and professional experience earned over time to see how that affected the employment of the applicants for different positions over time. There is little to no evidence that the phenomena we are investigating may change over time. The factors that we have found to increase chances of employment like whether the programmer has worked in a professional context before, how many languages they know, and what degree they have are set in stone necessities that will always be valued for jobs in the  computer science sector. 

## Future Directions and Next Steps
The lack of a timeline makes our analysis hard to pinpoint to a certain moment in time. The fast-paced nature of the Computer Science job market makes it so that even a year in difference could require a completely different toolkit of skills. As it is very well known, the CS job market is in an accelerating process of change where previous requirements, skillsets, and professional experiences are being abandoned in the wake of the importance of AI related skillsets. Then, maybe a succeeding study should focus on the qualitative qualities of skills rather than the technical ones. As our research indicates that Computer Skills and the quantity of them are the most important in securing a job, it would be fundamental to explore what, in the contemporary CS job market, can be defined as a Computer Skill. Our dataset suffers from 1-dimensionality. In job application forms, skills remain in this form, but our analysis isn’t able to cover the succeeding stages where applicants turn these words into pictures that describe themselves in relation to these skills.
