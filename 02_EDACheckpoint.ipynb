# Research Question

The main question we aim to answer: What factors most affect the likelihood of landing a job in the computer science field? Specifically, to what extent do the level of their education, whether or not the prospective applicant has prior experience, their gender, or their age increase their chances of employment in the computer science field?


## Background and Prior Work

The advent of the digital age brought upon the rise of the tech industry. Promises of job security, high pay, and comfortable working conditions within this field lured many into technology related careers. If anything, these figures help paint a picture of the outlook of the contemporary Computer Science prospect: studying within this field to secure a stable career, but faced with extreme competition, and high unemployment rates. Then, our research aims to determine what factors lie as an outlier within people employed within the Computer Science field – what factors best help someone realize the opportunity of getting a job?

An opinion piece published by Southern New Hampshire University claims that factors that help someone get a career in Computer Science include programming skills, critical thinking, and real-world experience. According to the U.S. Bureau of Labor Statistics, the typical entry-level education requirement is a Masters Degree. This figure also emphasizes the importance of analytical, problem-solving, and other career related skills. It can be inferred that one might use previous working experience in order to prove these skills.

In the article itself, Borden interviews a college student who gained internship experience, who claims that the internship helped reinforce skills required for her career.

A 2023 study by Caesar Jude Clemente developed predictive models to forecast the chances of Information Technology graduates finding a job. By surveying graduates from the United States and Canada, the research looked at how academic performance, socioeconomic status, study habits, and demographic data impact employability. The study identified previous experience, final and last semester GPA, and IT related project research as some of the most influential predictors for securing employment. This study emphasizes the importance of practical experience in securing a job.

A study by Felix L. Huerte Jr, et al. surveyed 488 computing studies alumni and used a decision tree algorithm to predict how long they would spend waiting for a job post-grad. The algorithm evaluated variables like age, gender, GPA, and internship experience. With an accuracy of 81%, the findings revealed that internship experience and academic performance were the strongest predictors of faster employment, while demographic factors like age and gender had minimal influence.

The dataset focuses on predicting whether a computer science job applicant is likely to be hired given features like education level, gender, prior experience and technical skills. Key metrics and How we read them

- Employment Outcome:- This is the main variable examined, that indicates whether the applicant was hired or not. In this project, we aim to examine how different variables relate to employability in the computer science field.
- Age:- Age might capture the difference in career stage and experience . Since our dataset has only 2 categories (below and above 35), we should interpret this metric carefully, avoiding drawing any overly specific conclusions.
Education Level:- Education level is one of the primary metrics of our analysis because we hypothesize that higher education often leads to better job outcomes. Differences in education levels may reflect the increasing importance of specialization and technical knowledge.
- Gender:-  We include gender to investigate if there exists any bias or discrimination in hiring processes. Differences observed across gender groups may reflect broader social inequalities  and systemic biases already prevalent in society.
- Years of Professional Coding Experience :- This metric represents professional industry experience. We use this variable as a proxy for technical ability. Higher values usually indicate greater industry experience and familiarity with real world tools. We expect candidates with more years of experience to show higher employment rates.
- Computer Skills:- The number of technical skills known by an applicant may reflect the breadth of the applicant's technical knowledge. Candidates with a larger number of skills may appear more well rounded to employers, though possessing one skill does not necessarily indicate domain expertise. We aim to examine how technical exposure relates to employment outcomes.


Our dataset consists of 70k+ job applicants and factors relating to their employability. Data is said to have been collected from a compilation of differing sources, within a specific period of time. However, neither sources nor specific timeframe are indicated within the dataset. We identify these characteristics as concerning because of the potential biases; the lack of timeframe and source specificity within the dataset means that the data being collected could be representative of an extrema. For example, if the data was collected in a time where hiring for Computer Science related roles was at an all time high, then the corresponding data would be misrepresentative of the overall Computer Science job market. The lack of specificity of sources indicates a similar problem. Data could be collected from extremely selective companies, which would then lead to data being representative of these high standards.

Another major issue is this dataset's bias towards sampling men. Of the 70k+ applicants, 93% identify as male, with 5% identifying as a woman and 2% identifying as other. There are a multitude of reasons why this could be the case, though in general, it seems to be representative of the general misogyny within technical roles. The imbalance within sampling indicates that data in the dataset isn’t going to be as representative of the overall application experience as it is for people from different backgrounds.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

cat_cols = ['EdLevel', 'Gender', 'Age']

fig, axes = plt.subplots(1, len(cat_cols), figsize=(20, 8))
fig.suptitle('Employment Rates by Categorical Variables', fontsize=18)

for i, col in enumerate(cat_cols):
    crosstab_prop = pd.crosstab(df[col], df['Employed'], normalize='index')

    if 1 in crosstab_prop.columns:
        crosstab_prop = crosstab_prop.sort_values(by=1, ascending=True)

    crosstab_prop.plot(kind='barh', stacked=True, ax=axes[i], color=['#ff9999', '#66b3ff'], width=0.7)

    axes[i].set_title(f'Proportion of Employed vs.\nUnemployed by {col}', fontsize=14)
    axes[i].set_xlabel('Proportion (0 to 1)')
    axes[i].set_ylabel(col)

    axes[i].legend(title='Status',
                   labels=['0: Unemployed', '1: Employed'],
                   loc='upper center',
                   bbox_to_anchor=(0.5, -0.15),
                   ncol=2)

    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.2f', label_type='center', color='black', weight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

The leftmost chart indicates that, within the dataset, having a higher education level doesn't necessarily correlate with being employed. In fact, this chart reveals that people without any higher education and people with only an undergraduate degree have the highest rates of unemployment.

The middle chart shows us the rates of employment between reported gender groups within our dataset. The chart indicates that men and nonbinary people have a noticeably higher rate of employment than women. It is important to note here, as mentioned above, that our data set is highly skewed towards male respondents (93%), and as such, the gender proportions for women and nonbinary people are much more prone to variance due to smaller sample sizes.

Finally, the rightmost chart shows us that younger applicants (under 35) have a slightly higher employment rate than older applicants. It should be duly noted that the dataset contains more participants under the age of 35 than participants older thatn 35 (65% and 35%, respectively.)

Overall, these three charts allow us to see existing disparites within the data, relative to each variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

employed_devs = df[df["Employed"] == True]
unemployed_devs = df[df["Employed"] == False]

numeric_cols = ["Age", "YearsCode", "YearsCodePro", "PreviousSalary", "ComputerSkills"]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for ax, col in zip(axes, numeric_cols):

    if col == "Age":
        sns.histplot(
            data=employed_devs,
            x=col,
            stat="density",
            alpha=0.55,
            color="steelblue",
            label="Employed",
            ax=ax,
            shrink=0.4,
        )
        sns.histplot(
            data=unemployed_devs,
            x=col,
            stat="density",
            alpha=0.55,
            color="tomato",
            label="Unemployed",
            ax=ax,
            shrink=0.4,
        )
    else:
        ax.hist(
            employed_devs[col],
            bins=30,
            alpha=0.55,
            color="steelblue",
            label="Employed",
            density=True,
        )
        ax.hist(
            unemployed_devs[col],
            bins=30,
            alpha=0.55,
            color="tomato",
            label="Unemployed",
            density=True,
        )

    ax.set_title(f"Distribution of {col}", fontsize=12)
    ax.set_ylabel("Density")
    ax.legend(fontsize=9)

fig.delaxes(axes[5])

plt.suptitle("How Behavioral Features Differ: Employed vs Unemployed Developers")
plt.tight_layout()
plt.show()
```

From the 5 histograms, we observe some patterns for each of the numeric variables among the employed and unemployed populations:

1. Age - We see that those who are employed make up a small majority of the 'under 35' age group and those who are unemployed make up a small majority of the 'over 35' age group. However, under both age groups, there is an almost equal split of employed and unemployed individuals, meaning that in our scenario, the age of an individual, by itself, is a poor predictor of employment.

2. YearsCode - For the entire range of YearsCode values, each bar is largely purple meaning that for each value of number of years of coding experience, there is significant overlap between the employed and unemployed classes. While the unemployed population is slightly greater when number of coding years is very small and very large and the employed population is slightly greater when number of coding years is in the middle of the range, these difference are minute, meaning that the number of employed and unemployed individuals for any specific number of coding years group is roughly equal. Hence, by itself, number of coding years is a poor predictor of employment.

3. YearsCodePro - Follows a similar trend compared to the YearsCode variable. Each bar is largely purple for every value of YearsCodePro, meaning that for any value of YearsCodePro, there is significant overlap between unemployed and employed classes, meaning that there is a roughly equal number of individuals in both classes for each of the values of YearsCodePro. Hence, by itself, YearsCodePro is a poor predictor of employment.

4. PreviousSalary - Similar trend compared to the previous variables. Since there is huge overlap of the two classes for each of the values of previous salary, for each salary group, there is a roughly equal number of employed and unemployed individuals. Hence, by itself, an individual's previous salary is a poor predictor of employment.

5. ComputerSkills - This is the only variable where the unemployed and employed classes have very little overlap. For the portion of this graph where number of skills is low, we see a rising density of unemployed individuals while the number of employed individuals is negligible in here, indicating that if someone has less computer skills, it is almost certain that they will not be hired. Towards the middle portion of the graph, there is some overlap between the unemployed and employed classes as in this portion, the density of unemployed individuals falls while the density of employed individuls rise. Hence, for this specific range of computer skills, there is a nearly equal chance of being employed or unemployed. Finally, in the portion of the graph with an even higher number of computer skills, while the density of employed individuals does decline, the density of unemployed individuals is negligible (nearly 0). Since there is once again a negligible amount of overlap between the two classes, this indicates that if an individual possesses a high number of computer skills, it is almost certain that they will be employed.

Overall, from our 5 histograms that attempt to assess the relationship between being employed and each of the 5 numeric variables, there appears to be a meaningful relationship only with the number of computer skills. The distribution of YearsCodePro is right-skewed, indicating that most candidates have fewer years of professional experience. While there seems to be a small correlation between employment outcomes and higher professional experience, there still seems to be a significant overlap between two groups. This suggests that years of professional coding experience does not strongly affect employment outcomes.

**Previous Salary**:
The scatterplot between previous salary and years of professional coding experience show a positive relationship. This is expected as experienced developers usually earn higher salaries. However, there again seems to be a significant overlap between employed and unemployed candidates across all salary levels. This suggests that previous salary does not strongly affect employment outcomes.

**Computer Skills**:
Computer skills seems to show the strongest correlation with employment outcomes. Employed candidates seem to cluster at a higher numbers of technical skills, while the unemployed candidates seem to have lower number of technical skills. This aligns with our logistic regression results below, where computer skills emerged as the strongest predictor of employment outcomes, proudcing the largest psedo R² value during forward selection.

```python
#Removed Country and YearsCode because they weren't as relevant and Employed since that is the variable we are predicting
columns_of_interest = ['Age', 'C(EdLevel)', 'C(Gender)', 'C(MainBranch)',
       'YearsCodePro', 'PreviousSalary', 'ComputerSkills']
previous_features = [] #start with null model for feature selection
best_r2 = 0.0 # best r-squared seen so far, and by definition a null model has no r squared

while len(columns_of_interest) > 0:
  pseudor2 = [] #a list to keep track of all the models we will try
  for col in columns_of_interest:
      current_features = previous_features + [col]
      indep_vars = ' + '.join(current_features)
      formula = 'Employed ~ {}'.format(indep_vars)

      a_model = smf.logit(formula, data=df).fit()
      print('fitting ', formula, '\t\t R2: {:5.4f}'.format(a_model.prsquared) )
      pseudor2.append( (indep_vars, a_model.prsquared) )

  results = pd.DataFrame.from_records(pseudor2, columns=['independent variables', 'R2']).sort_values(by='R2', ascending=False)

  # here we are setting ourselves up for the next step in model selection
  selected = results['independent variables'].iloc[0] # winning formula is the first one in the sorted R2 list
  previous_features = selected.split(' + ') # remove the formula stuff, leave us with a list of predictors from the winning model of this step
  columns_of_interest.remove( previous_features[-1] ) # remove the new predictor we just added to the formula from the list of predictors to consider next time
  new_r2 = results['R2'].iloc[0] # best pseudo R2 of the step...
  # new_r2 is useful for comparing to best_r2 to see if we've met a stopping condition for forward selection

  print('\n on this step we selected the model: {} \t\t R2: {:5.4f}\n vs previous best R2: {:5.4f}'.format(selected, new_r2, best_r2))
  formula = 'Employed ~ {}'.format(selected)
  a_model = smf.logit(formula, data=df).fit()
  print(a_model.summary())
```

**Forward Selection using Logistic Regression**

To identify which variables most strongly predict employment outcome, we implemented a forward selection model using logistic regression. Since our target variable was a binary categorical variable, logistic regression was the most appropriate choice.

The variables considered for the model were:-

- Age
- Education Level
- Gender
- MainBranch
- Years of Professional coding experience
- Previous Salary
- Computer Skills

We started with a null model, we iteratively added variables that produced the largest increase in McFadden pseudo R² at each step. Models were ranked by their R² value, and the best-performing model was added permanently before repeating the process.

**Findings**

The forward model shows us that ComputerSkills was by far the strongest predictor of employment. When added alone, it increased the model's pseudo R² to around 0.35, while the other variables each produced pseudo R² values close to 0. These findings suggest that the breadth of technical skills is the distinguishing factor between the employed and unemployed applicants within our dataset. These results are consistent with the visualizations shown above.
