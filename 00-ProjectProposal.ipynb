## Research Question

The main question we aim to answer: What factors most affect the likelihood of landing a job in the computer science field? Specifically, to what extent do the level of their education, whether or not the prospective applicant has prior experience, their gender, or their age increase their chances of employment in the computer science field?

## Background and Prior Work

The advent of the digital age brought upon the rise of the tech industry. Promises of job security, high pay, and comfortable working conditions within this field lured many into technology related careers. The U.S. News & World Report‘s 100 Best Jobs list, for instance, lists 4 technology adjacent jobs in their top 10 <a name="back1"></a><sup>[1](#ref1)</sup>. These positions rank higher than other established careers like Lawyer, Psychiatrist, etc. The aforementioned promises, rankings such as the one above, and other factors catalyzed a surge of students studying Computer Science. From 2013 to 2022, the number of students studying towards a degree in Computer Science doubled from 51,696 to 112,720 <a name="back2"></a><sup>[2](#ref2)</sup>. In fact, from this same figure, it can be seen that Computer Science degrees have the largest increase among all other fields. 

Unfortunately, the market has failed to keep up with demand. Across the United States, as of 2024, the unemployment rate for recent graduates studying Computer Science is 7% <a name="back3"></a><sup>[3](#ref3)</sup>. In this figure, Computer Science majors lie within the top 5 unemployed majors when ranked by unemployment. Furthermore, the unemployment rate lies 2.8% higher than the national average. This pattern can also be observed in specific careers within the Computer Science field. Take, for instance, the software engineer position, where employment has dropped by 27.9% since 2019, as per ADP Research <a name="back4"></a><sup>[4](#ref4)</sup>. 

If anything, these figures help paint a picture of the outlook of the contemporary Computer Science prospect: studying within this field to secure a stable career, but faced with extreme competition, and high unemployment rates. Then, our research aims to determine what factors lie as an outlier within people employed within the Computer Science field – what factors best help someone realize the opportunity of getting a job? 

An opinion piece published by Southern New Hampshire University claims that factors that help someone get a career in Computer Science include programming skills, critical thinking, and real-world experience <a name="back5"></a><sup>[5](#ref5)</sup>. According to the U.S. Bureau of Labor Statistics, the typical entry-level education requirement is a Masters Degree <a name="back6"></a><sup>[6](#ref6)</sup>. This figure also emphasizes the importance of analytical, problem-solving, and other career related skills. It can be inferred that one might use previous working experience in order to prove these skills. 

In fact, a 2024 post by Taylor Borden, an editor at LinkedIn News, uses data from LinkedIn to prove that new grads were 22.9% more likely to start a full-time job within 6 months of graduation if they had internship experience <a name="back7"></a><sup>[7](#ref7)</sup>. In the article itself, Borden interviews a college student who gained internship experience, who claims that the internship helped reinforce skills required for her career.

## Hypothesis

This is based on our prior notions that internships have become increasingly important in today’s society to bolster an applicant’s chance of employment and that a bachelor’s degree is no longer a guarantee of employment, and a masters or graduate degree may be needed to compete with others in such a competitive field. We look to find how these developments have affected the market, especially since three of us are Computer Science majors.

## Ethics 

1. Data Collection and Privacy
We are sourcing our stats from public reports like the BLS, ADP, and aggregate LinkedIn data. Since we aren't scraping individual profiles or handling personally identifiable information (PII) ourselves, the privacy risks on our end are pretty low. However, we are relying on the assumption that platforms like LinkedIn and ADP aggregated their users' data ethically and with proper consent before publishing their findings.

2. Bias and Representation in Data
There is a definite risk of representation bias in our sources. For example, the LinkedIn data only accounts for people who actively use the platform to network and update their employment status. This potentially skews the numbers, leaving out recent grads who found jobs through other avenues or who don't maintain an online professional presence. Additionally, highlighting internship experience as a primary success factor might overlook socioeconomic barriers; unpaid or underpaid internships are often only accessible to more privileged students, which could skew our understanding of who is actually securing these entry-level roles.

3. Transparency and Accountability
To keep things transparent, we’ve clearly cited all of our sources using inline footnotes and direct links so anyone reading our notebook can verify the numbers. If our conclusions end up being off base or the market suddenly shifts, other researchers can easily trace back exactly where we got our initial data.

4. Potential for Harm or Unintended Consequences
We need to be careful about how we frame the current "doom and gloom" of the CS job market. If our research is taken out of context, it could unnecessarily panic current students or discourage people from pursuing tech entirely, even though the job market is cyclical. Our goal is to highlight the reality of the competition and the value of experience, without making it sound like getting a job is impossible.
